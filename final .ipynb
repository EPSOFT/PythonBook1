```python
class Animal:
    def __init__(self, name):
        self.name = name
    
    def sound(self):
        return "Some sound"  # روش عمومی

class Dog(Animal):  # وراثت از Animal
    def sound(self):  # بازنویسی روش (چندشکلی)
        return "Woof!"

class Cat(Animal):
    def sound(self):
        return "Meow!"

# ساخت اشیا
my_dog = Dog("Buddy")
my_cat = Cat("Whiskers")

print(my_dog.sound())  # خروجی: Woof!
print(my_cat.sound())  # خروجی: Meow!

```

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.__balance = balance  # خصوصی (کپسوله‌سازی)
    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
        return self.__balance
    
    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
        return self.__balance
    
    def get_balance(self):  # getter برای دسترسی ایمن
        return self.__balance

# ساخت شی
account = BankAccount("Alice", 100)
print(account.deposit(50))  # خروجی: 150
print(account.withdraw(20))  # خروجی: 130
# print(account.__balance)  # خطا: دسترسی مستقیم ممکن نیست (هرچند در پایتون قابل هک است)
```

```python
def safe_divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError as e:
        print("خطا: تقسیم بر صفر مجاز نیست.")
        return None
    else:
        print("تقسیم موفق:", result)
        return result
    finally:
        print("پایان تابع safe_divide — پاکسازی (در صورت نیاز)")
        
print(safe_divide(10, 2))
print(safe_divide(5, 0))

```

```python
import requests

def fetch_data(url):
    try:
        response = requests.get(url, timeout=5)
        response.raise_for_status()  # بررسی خطاهای HTTP
    except requests.Timeout:
        print("خطا: زمان درخواست به پایان رسید.")
        return None
    except requests.HTTPError as e:
        print(f"خطای HTTP: {e}")
        return None
    except requests.ConnectionError:
        print("خطا: مشکل در اتصال به شبکه.")
        return None
    else:
        return response.json()

# تست
data = fetch_data("https://api.example.com/data")
print(data)
```

```python
def read_file_safely(filepath):
    try:
        with open(filepath, 'r', encoding='utf-8') as f:
            return f.read()
    except FileNotFoundError:
        print(f"خطا: فایل '{filepath}' یافت نشد.")
        return None
    except PermissionError:
        print(f"خطا: دسترسی به فایل '{filepath}' ممکن نیست.")
        return None

# تست
print(read_file_safely('nonexistent.txt'))
print(read_file_safely('example.txt'))
```

```python
from pathlib import Path

# ایجاد شیء Path برای یک فایل یا پوشه
p = Path('file.txt')

# تغییر نام فایل یا انتقال آن به مسیر دیگر
p.rename('new_file.txt')          # تغییر نام یا انتقال (حرکت) به مسیر جدید

# حذف فایل (از پایتون ۳.۸ به بعد)
p.unlink(missing_ok=True)         # اگر فایل وجود نداشته باشد خطا نمی‌دهد

# ایجاد پوشه
Path('folder').mkdir(exist_ok=True)               # ایجاد پوشه (اگر وجود داشته باشد خطا نمی‌دهد)
Path('a/b/c').mkdir(parents=True, exist_ok=True)  # ایجاد مسیر چندسطحی (parents=True)

# لیست کردن محتوای پوشه
list(Path('.').iterdir())                         # تمام فایل‌ها و پوشه‌های داخل پوشه فعلی

# بررسی وجود و نوع مسیر
p.exists()                                        # آیا فایل/پوشه وجود دارد؟
p.is_file()                                       # آیا یک فایل است؟
p.is_dir()                                        # آیا یک پوشه است؟

# اطلاعات فایل (مثل اندازه)
p.stat().st_size                                  # اندازه فایل به بایت
p.stat().st_mtime                                 # زمان آخرین تغییر (timestamp)

# جستجوی فایل‌ها (با الگوی glob)
list(Path('.').glob('*.log'))                     # تمام فایل‌های .log فقط در پوشه فعلی
list(Path('.').rglob('*.py'))                     # تمام فایل‌های .py به صورت بازگشتی در تمام زیرپوشه‌ها
```

```python

import csv
from pathlib import Path

# داده‌های نمونه
data = [
    ["نام", "سن", "شهر"],
    ["خشایار", 25, "تهران"],
    ["آتوسا", 30, "اصفهان"],
    ["کوروش", 22, "شیراز"]
]

# نوشتن در فایل CSV (روش استاندارد و بدون مشکل)
csv_path = Path('people.csv')

with csv_path.open('w', encoding='utf-8', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(data)

print(f"فایل با موفقیت در {csv_path.resolve()} ذخیره شد.")

# خواندن از فایل CSV (روش تمیز و حرفه‌ای)
with csv_path.open('r', encoding='utf-8', newline='') as f:
    reader = csv.reader(f)
    
    # خواندن سرستون
    headers = next(reader)
    print("سرستون‌ها:", headers)
    
    print("داده‌ها:")
    for row in reader:
        print(f"نام: {row[0]}, سن: {row[1]}, شهر: {row[2]}")
```

```python
def read_file_safely(filepath):
    try:
        with open(filepath, 'r', encoding='utf-8') as f:
            return f.read()
    except FileNotFoundError:
        print(f"خطا: فایل '{filepath}' یافت نشد.")
        return None
    except PermissionError:
        print(f"خطا: دسترسی به فایل '{filepath}' ممکن نیست.")
        return None

# تست
print(read_file_safely('nonexistent.txt'))
print(read_file_safely('example.txt'))
```

```python
import requests

def get_stock_data():
    url = "https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol=IBM&interval=5min&outputsize=full&apikey=demo"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        last_refreshed = data["Meta Data"]["3. Last Refreshed"]
        price = data["Time Series (5min)"][last_refreshed]["1. open"]
        return price
    else:
        return None

price = get_stock_data()
symbol = "IBM"
if price is not None:
    print(f"{symbol}: {price}")
else:
    print("Failed to retrieve data.")

```

```python
import json
import requests

def fetch_and_print_articles(api_url):
    response = requests.get(api_url)
    
    if response.status_code == 200:
        articles = response.json().get('articles', [])
        for index, article in enumerate(articles[:1], start=1):
            print(f"Article {index}:\n{json.dumps(article, sort_keys=True, indent=4)}\n")
    else:
        print(f"Error: {response.status_code}")

API_KEY = 'paste_your_api_key_here'
api_endpoint = f"https://newsapi.org/v2/top-headlines?country=us&category=business&apiKey={API_KEY}"

fetch_and_print_articles(api_endpoint)
```

```python
try:
    my_list = [1, 2, 3]
    print(my_list[5])  # IndexError
except LookupError as e:
    print(f"خطای جستجو: {e}")
except Exception as e:
    print(f"خطای عمومی: {e}")
```

```python
def safe_divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError as e:
        print("خطا: تقسیم بر صفر مجاز نیست.")
        return None
    else:
        print("تقسیم موفق:", result)
        return result
    finally:
        print("پایان تابع safe_divide — پاکسازی (در صورت نیاز)")

print(safe_divide(10, 2))
print(safe_divide(5, 0))
```

```python
def set_age(age):
    if age < 0:
        raise ValueError("سن نمی‌تواند منفی باشد")
    return age

try:
    set_age(-5)
except ValueError as e:
    print("دریافت شد:", e)
```

```python
class ValidationError(Exception):
    """خطای اعتبارسنجی برای داده‌های ورودی"""
    pass

def validate_username(name):
    if not name:
        raise ValidationError("نام کاربری خالی است")
    if len(name) < 3:
        raise ValidationError("نام کاربری باید حداقل 3 حرف باشد")
    return True

try:
    validate_username('ab')
except ValidationError as e:
    print('خطای اعتبارسنجی:', e)
```

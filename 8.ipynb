```python
def is_even(n):
    return n % 2 == 0
print(is_even(4))  # خروجی: True
```

```python
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a
print(gcd(48, 18))     #6:خروجی
```

```python
s = "1234"# اعداد فارسی    
print(s.isdecimal()) # خروجی : True
print(s.isnumeric())  # خروجی :True
print(s.isdigit())  # خروجی :True
```

```python
s2 = "Ⅷ"   # (8) عدد رومی
print(s2.isdecimal())  # خروجی: False
print(s2.isnumeric())  # خروجی: False
print(s2.isdigit())    # خروجی: False
```

```python
s = "Python"
b = s.encode("utf-8")   # تبدیل به بایت
print(b)                # خروجی: b'Python'

s2 = b.decode("utf-8")  # تبدیل بایت به رشته
print(s2)               # خروجی: Python
```

```python
# نوشتن در فایل
with open('example.txt', 'w', encoding='utf-8') as f:
    f.write('سلام دنیا\n')
    f.write('این یک خط دیگر است.\n')

# خواندن کامل فایل
with open('example.txt', 'r', encoding='utf-8') as f:
    content = f.read()
print(content)

# خواندن خط به خط
with open('example.txt', 'r', encoding='utf-8') as f:
    for i, line in enumerate(f, start=1):
        print(f"خط {i}: {line.strip()}")
```

```python
# نوشتن داده باینری
data = bytes([0x00, 0xFF, 0x7A])
with open('example.bin', 'wb') as f:
    f.write(data)

# خواندن بایت‌ها
with open('example.bin', 'rb') as f:
    b = f.read()
print(b)
```

```python
import os

# تغییر نام فایل
os.rename('example.txt', 'renamed_example.txt')

# بررسی وجود و حذف
if os.path.exists('renamed_example.txt'):
    print('فایل وجود دارد')
    os.remove('renamed_example.txt')

# اطلاعات فایل
stat = os.stat('example.bin')
print('حجم فایل:', stat.st_size)
```
